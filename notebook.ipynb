## G. Statistical Inference (ANOVA)

````python
from scipy.stats import f_oneway
import pandas as pd

# Membuat kategori impressions
df['impressions_category'] = pd.qcut(
    df['Impressions'],
    q=3,
    labels=['Low', 'Medium', 'High']
)

# Variabel yang akan diuji
anova_variables = [
    'Likes',
    'Shares',
    'Comments',
    'Saves',
    'From Hashtags'
]

# Loop ANOVA untuk setiap variabel
for variable in anova_variables:
    
    low_group = df[df['impressions_category'] == 'Low'][variable]
    medium_group = df[df['impressions_category'] == 'Medium'][variable]
    high_group = df[df['impressions_category'] == 'High'][variable]
    
    anova_result = f_oneway(
        low_group,
        medium_group,
        high_group
    )
    
    print(f'ANOVA Result for {variable}')
    print('F-statistic :', anova_result.statistic)
    print('P-value :', anova_result.pvalue)
    print('-'*40)
````

## Interpretasi Hasil Uji ANOVA

Berdasarkan hasil uji ANOVA, seluruh variabel engagement memiliki nilai p-value kurang dari 0.05. Hal ini menunjukkan bahwa H0 ditolak dan terdapat perbedaan signifikan rata-rata engagement berdasarkan kategori impressions (Low, Medium, dan High).

### 1. Likes
Variabel Likes memiliki nilai F-statistic sebesar 51.13 dengan p-value 5.59e-16.

Hasil ini menunjukkan bahwa jumlah likes berbeda secara signifikan pada setiap kategori impressions. Konten dengan impressions tinggi cenderung memperoleh likes lebih banyak dibandingkan konten dengan impressions rendah.

### 2. Shares
Variabel Shares memiliki nilai F-statistic sebesar 9.11 dengan p-value 0.00023.

Hal ini menunjukkan bahwa jumlah shares juga berbeda signifikan antar kategori impressions. Konten dengan jangkauan lebih luas memiliki peluang lebih besar untuk dibagikan oleh pengguna.

### 3. Comments
Variabel Comments memiliki nilai F-statistic sebesar 5.02 dengan p-value 0.0084.

Meskipun hubungan comments tidak sekuat likes atau saves, hasil ini tetap menunjukkan bahwa jumlah komentar memiliki perbedaan signifikan berdasarkan kategori impressions.

### 4. Saves
Variabel Saves memiliki nilai F-statistic sebesar 20.17 dengan p-value 4.49e-08.

Hal ini menunjukkan bahwa konten dengan impressions tinggi cenderung lebih banyak disimpan oleh pengguna. Saves menjadi indikator bahwa konten dianggap menarik atau memiliki value bagi audiens.

### 5. From Hashtags
Variabel From Hashtags memiliki nilai F-statistic sebesar 35.54 dengan p-value 2.33e-12.

Hasil ini menunjukkan bahwa distribusi konten dari hashtags memiliki perbedaan signifikan antar kategori impressions. Penggunaan hashtag yang efektif dapat membantu meningkatkan jangkauan konten Instagram Reels.

## Kesimpulan

Secara keseluruhan, hasil ANOVA menunjukkan bahwa impressions memiliki hubungan signifikan terhadap berbagai bentuk engagement seperti likes, shares, comments, saves, dan distribusi dari hashtags.
